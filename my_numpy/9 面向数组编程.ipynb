```python
#numpy数组可以将许多种数据处理任务表述为简洁的数组表达式,而无需编写循环
#使用数组表达式代替循环的过程称为  "向量化"

import numpy as np

#例子:我们想在网格数据上计算sqrt(x**2+y**)
#方法:np.meshgrid(arr_x,arr_y)接受两个一维数组,返回两个二维数组


#创建一维网格坐标
points=np.arange(-5,5,0.01)
points.ndim #1维

#创建二维网格坐标
xs,ys=np.meshgrid(points,points)
print(f"{ys}\n{xs}") #2维

"""
xs:所有点的 x 坐标组成的矩阵。
ys:所有点的 y 坐标组成的矩阵。
"""
```

```python
#使用数组当作2个浮点数编写进表达式即可:

sqrt_arr=np.sqrt(xs**2+ys**2)#形成2维数组
print(sqrt_arr)
```

```python
#可视化
#使用matplotlib可视化这个二维数组
import matplotlib.pyplot as plt

image=plt.imshow(sqrt_arr,cmap=plt.cm.gray,extent=[-5,5,-5,5])

#添加颜色条(显示数值与颜色的对应关系)和标题:
plt.colorbar(image)
plt.title("image plot of $\sqrt{x^2+y^2}$ for a grid of values")

#添加横纵坐标:
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
