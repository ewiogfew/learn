```python
import numpy as np
import matplotlib.pyplot as plt

# 
nwalks = 100  # 
nsteps = 1000   # 

#
rng=np.random.default_rng(seed=12345)
draws = 0.25 * rng.standard_normal((nwalks, nsteps))

# 
draws[:, 0] = 100 

# 
walks=draws.cumsum(1)

# 
plt.plot(walks)
plt.title('Random Walks')
plt.xlabel('Step')
plt.ylabel('Value')
plt.show()
```
