```python
import numpy as np
import matplotlib.pyplot as plt

# 参数定义
nwalks = 100  # 漫步次数
nsteps = 1000   # 步数

#生成标准正态分布数据
rng=np.random.default_rng(seed=12345)
draws = 0.25 * rng.standard_normal((nwalks, nsteps))

# 设置初始值
draws[:, 0] = 100 

# 进行漫步
walks=draws.cumsum(1)

# 绘制漫步过程
plt.plot(walks)
plt.title('Random Walks')
plt.xlabel('Step')
plt.ylabel('Value')
plt.show()
```
